```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=1000, centers=3, n_features=2,
                  random_state=0)
```

```python
model = DBSCAN(min_samples=35, eps=0.5).fit(X)
labels = model.labels_
```

```python
colors = {} 
colors[0] = 'r'
colors[1] = 'g'
colors[2] = 'b'
colors[3] = 'c'
colors[4] = 'y'
colors[5] = 'm'
colors[6] = 'orange'
colors[7] = 'pink'
colors[8] = 'brown'
colors[-1] = 'k'

cvec = [colors[label] for label in labels]

plt.scatter(X[:,0], X[:,1], c = cvec)
plt.show()
```
